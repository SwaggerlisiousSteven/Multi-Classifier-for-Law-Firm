```python
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from nltk.corpus import stopwords
import itertools, string, operator, re, unicodedata, nltk
from operator import itemgetter
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import  RegexpTokenizer
import numpy as np
from itertools import combinations
from gensim.models import Phrases
from collections import Counter
from nltk.corpus import words
from gensim.parsing.preprocessing import STOPWORDS

words = set(nltk.corpus.words.words())
punc = list(set(string.punctuation))
#my_stop_words = STOPWORDS.union(set(['court','case','section','order','said', 'made']))
#stopwords = stopwords.words('english')
#stop= stopwords.extend(['court','case','section','order','said', 'made'])
#new_stopwords= ['court','case','section','order','said', 'made']
remove_words = ['court','case','section','order','said', 'made', 'would', 'also']
#new_stopwords_list = stopwords.union(new_stopwords)
df1['K_Content'] = df1['K_Content'].astype(str)

def remove_punct(text):
    no_punct = "".join([c for c in text if c not in punc])
    return no_punct
def remove_singleletter(text):
    re_letter= ' '.join( [w for w in text.split() if len(w)>3] )
    return re_letter
def remove_nonenglish_words(text):
    return " ".join(w for w in nltk.wordpunct_tokenize(text) 
     if w.lower() in words or not w.isalpha())
#def repeat_words(text):
    #ulist = ['court']
    #[ulist.append(y) for y in text if y not in ulist]
    #return ulist
def repeat_words(text):
    new_words = " ".join([word for word in text.split() if word.lower() not in remove_words])
    return new_words


df1['content_punct'] = df1['K_Content'].apply(lambda x: remove_punct(x))
df1['content_punct']= df1['content_punct'].apply(lambda x: x.lower())
df1['content_punct']=df1['content_punct'].str.replace('\d+', '')
df1['content_punct']= df1['content_punct'].apply(lambda x: remove_singleletter(x))
df1['content_punct']= df1['content_punct'].apply(lambda x: remove_nonenglish_words(x))
df1['content_punct']= df1['content_punct'].apply(lambda x: repeat_words(x) )
tokenizer= RegexpTokenizer(r'\w+')
df1['content_punct']=df1['content_punct'].apply(lambda x: tokenizer.tokenize(x.lower()))


def remove_stopwords(text):
    #words= [w for w in text if w not in stopwords.extend(['court','case','section','order','said', 'made'])]
    words= [w for w in text if w not in stopwords.words('english')]
    #words= [w for w in text if w not in stop]
    return words
df1['content_punct']=df1['content_punct'].apply(lambda x: remove_stopwords(x))

#df1.head(20)
#df1.head(1)
df1['content_punct'][0]
```

```python
def word_freq(clean_text_list, top_n):
    """
    Word Frequency
    """
    flat = [item for sublist in clean_text_list for item in sublist]
    with_counts = Counter(flat)
    top = with_counts.most_common(top_n)
    word = [each[0] for each in top]
    num = [each[1] for each in top]
    return pd.DataFrame([word, num]).T

cl_text_list = df1['content_punct'].tolist()
wf = word_freq(cl_text_list, 50)
wf.head(50)
```
